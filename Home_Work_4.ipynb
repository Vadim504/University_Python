```python
import random

# 1. Создаем исходный файл со случайными числами
def create_input_file(filename, n):
    with open(filename, 'w') as f:
        for _ in range(n):
            f.write(f"{random.randint(1, 10)}\n")

# 2. Создаем файл с последовательными произведениями
def create_product_file(input_filename, output_filename):
    with open(input_filename, 'r') as infile, open(output_filename, 'w') as outfile:
        product = 1
        for line in infile:
            num = int(line.strip())
            product *= num
            outfile.write(f"{product}\n")

# 3. Параметры задачи
N = 10  # Количество чисел
input_file = "input_numbers.txt"
output_file = "output_products.txt"

# 4. Выполняем операции
create_input_file(input_file, N)
create_product_file(input_file, output_file)

# 5. Выводим результаты
print("Исходный файл:")
!cat {input_file}
print("\nРезультирующий файл:")
!cat {output_file}
```

```python
import random
from datetime import datetime, timedelta

def generate_random_date():
    # Генерируем случайную дату за последние 5 лет
    start_date = datetime.now() - timedelta(days=5*365)
    end_date = datetime.now()

    random_days = random.randint(0, (end_date - start_date).days)
    random_date = start_date + timedelta(days=random_days)

    # Форматируем строго в д.м.г
    return random_date.strftime("%d.%m.%Y")

def create_date_file(filename, n):
    with open(filename, 'w', encoding='utf-8') as f:
        for _ in range(n):
            date_str = generate_random_date()
            f.write(date_str + "\n")

def find_spring_dates(filename):
    spring_dates = []

    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if not line:  # Пропускаем пустые строки
                continue

            try:
                day, month, year = line.split('.')
                month = int(month)

                # Проверяем, что месяц весенний (3, 4 или 5)
                if 3 <= month <= 5:
                    spring_dates.append(line)
            except (ValueError, IndexError):
                # Пропускаем строки с некорректным форматом
                continue

    return spring_dates

N = 10
date_file = "dates.txt"
create_date_file(date_file, N)
spring_dates = find_spring_dates(date_file)

print(f"Содержимое файла '{date_file}':")
with open(date_file, 'r', encoding='utf-8') as f:
    print(f.read())

print("Весенние даты:")
for date in spring_dates:
    print(date)
```

```python
import random
import re

def generate_russian_like_word(length):
    russian_letters = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
    return ''.join(random.choice(russian_letters) for _ in range(length))

def create_file(filename, num_words=15, min_len=3, max_len=8):
    with open(filename, 'w', encoding='utf-8') as f:
        for _ in range(num_words):
            word_length = random.randint(min_len, max_len)
            word = generate_russian_like_word(word_length)
            f.write(word + ' ')


def get_sorted_words(filename):
    """Возвращает отсортированный список слов из файла"""
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()

    # Извлекаем слова (игнорируем пунктуацию)
    words = re.findall(r'\b[а-яё]+\b', text.lower())

    # Удаляем дубликаты и сортируем
    return sorted(set(words))
# Создаем файл
N = 20
filename = "words.txt"
create_file(filename, N)  # 20 слов

print(f"Содержимое файла '{filename}':")
with open(filename, 'r', encoding='utf-8') as f:
    print(f.read())

sorted_words = get_sorted_words(filename)
print("\nСлова в алфавитном порядке:")
for i, word in enumerate(sorted_words, 1):
    print(f"{i}. {word}")
```
