```python
def function(string):
  words = string.split()
  dict = {}

  for word in words:
    if word in dict:
      dict[word]+= 1
    else:
      dict[word]= 1
  return dict

string = "Apple Banana Apple Banana Orange"
function(string)
```

```python
def function(dict1,dict2):
  for key, value in dict2.items():
    if key in dict1:
        dict1[key] += value
    else:
        dict1[key] = value
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
print(function(dict1,dict2))
```

```python
def delete(dict1,list1):
  dict2 = dict1.copy()
  for key in list1:
      if key in dict2:
        del dict2[key]
  return dict2

dict1 = {1:1,2:3,3:5,4:7,5:9}
list1 = [1,2,3]
print(delete(dict1,list1))
```

```python
def function(dict1,dict2):
  result = {}
  all_keys = set(dict1.keys()) | set(dict2.keys())

  for key in all_keys:
      result[key] = []
      if key in dict1:
          result[key].append(dict1[key])
      if key in dict2:
          result[key].append(dict2[key])
  return result

dict1 = {1:1,2:3,3:5,4:7,5:9,6:11,7:13}
dict2 = {1:1,3:5,5:9,6:8,8:10}
print(function(dict1,dict2))
```

```python
def sort_dict(dict1):
  dict2 = dict(sorted(dict1.items()))
  return dict2

dict1 = {3:5,2:1,1:3,4:7,5:9}
dict2 = sort_dict(dict1)
for key in dict2:
  print(f"{key}: {dict2[key]}")

```

```python
from operator import itemgetter
def sort_dict(dict1):
  dict2 = dict(sorted(dict1.items(),key=itemgetter(1)))
  return dict2

dict1 = {3:5,2:1,1:3,4:7,5:9}
dict2 = sort_dict(dict1)
for key in dict2:
  print(f"{key}: {dict2[key]}")
```
