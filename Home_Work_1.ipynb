```python
def is_leap_year(year):
    return year % 400 == 0 or (year % 100 != 0 and year % 4 == 0)

def get_previous_day(g, m, n):
    if n > 1:
        return g, m, n - 1

    if m == 1:
        return g - 1, 12, 31

    if m == 3:
        if is_leap_year(g):
            return g, 2, 29
        else:
            return g, 2, 28

    if m in [5, 7, 10, 12]:
        return g, m - 1, 30
    else:
        return g, m - 1, 31

def get_next_day(g, m, n):
    if m in [1, 3, 5, 7, 8, 10, 12]:
        if n < 31:
            return g, m, n + 1
        else:
            if m == 12:
                return g + 1, 1, 1
            else:
                return g, m + 1, 1
    elif m in [4, 6, 9, 11]:
        if n < 30:
            return g, m, n + 1
        else:
            return g, m + 1, 1
    elif m == 2:
        if is_leap_year(g):
            if n < 29:
                return g, 2, n + 1
            else:
                return g, 3, 1
        else:
            if n < 28:
                return g, 2, n + 1
            else:
                return g, 3, 1

# Ввод данных
g = int(input())
m = int(input())
n = int(input())

# Определение предыдущего и следующего дня
prev_day = get_previous_day(g, m, n)
next_day = get_next_day(g, m, n)

# Вывод результатов
print(f"{prev_day[0]}-{prev_day[1]}-{prev_day[2]}")
print(f"{next_day[0]}-{next_day[1]}-{next_day[2]}")
```
